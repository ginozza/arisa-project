```python
from matplotlib import cm

X = fruits[['height', 'width', 'mass', 'color_score']]
y = fruits['fruit_label']

# Split given data into 2 sets. One for training and one for testing. Using the same data for training and 
# testing doesn't tell you anything new.
# A 75%-25% split is quite standard.
# random_state provides the seed value to the internal random number generator. 
#Providing the same value every time generates the same random number every time.
# X holds data, y holds label values.

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

cmap = cm.get_cmap('gnuplot')
scatter = pd.plotting.scatter_matrix(X_train, 
                                     c = y_train, marker = 'o', 
                                     s = 40, 
                                     hist_kwds = {'bins':15}, 
                                     figsize = (10, 10), cmap = cmap)
```
