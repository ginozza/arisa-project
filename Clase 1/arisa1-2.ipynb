```python
df_comp.spx.plot(figsize=(20,5), title = "S&P500 Prices")
plt.show()

# El eje x es la frecuencia (1 vez por dia), no aparece la fecha de momento, solo el indice del punto correspondiente
# Se alternan periodos de crecimiento y decrecimiento, se corresponde con la burbuja. COM (200) y la burbuja inmobiliaria (4000) 
```

```python
df_comp.ftse.plot(figsize = (20,5), title = "FTSE100 Prices")
plt.show()
# Sigue un patron similar al SP500 con las bajadas en 2000 y 2008
```

```python
df_comp.spx.plot(figsize = (20,5), title = "S&P500 Prices")
df_comp.ftse.plot(figsize = (20,5), title = "FTSE100 Prices")
plt.title("S&P vs FTSE")
plt.show()

# Dibujar series de forma conjunta
# Los valores de FTSE son mayores pero se mueven de manera similar - hay el mismo patron
# Esto NO sugiere que SP500 sea mas estable
# El analisis de tiempo en bolsa no solo mira precios sino retornos para evitar estas cosas
```
